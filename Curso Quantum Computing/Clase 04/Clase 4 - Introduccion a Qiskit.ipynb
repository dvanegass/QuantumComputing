### ¿Qué es Qiskit

Qiskit es un kit de desarrollo de software (SDK) de código abierto para la computación cuántica. Fue desarrollado por IBM y lanzado en 2017, con el objetivo de hacer más fácil para investigadores, educadores y desarrolladores trabajar con computadoras cuánticas y desarrollar algoritmos cuánticos.

#### Componentes de Qiskit

Qiskit está compuesto por varios componentes, incluyendo:

*   **Terra**: Terra es la base de Qiskit y proporciona las herramientas para el diseño, compilación y ejecución de circuitos. Incluye un conjunto de APIs para crear y manipular circuitos cuánticos, así como un compilador que optimiza los circuitos para diferentes tipos de hardware cuántico.

*   **Aer**: Aer es un simulador de alto rendimiento incluido en Qiskit que puede simular circuitos cuánticos en computadoras clásicas. Proporciona varios métodos de simulación diferentes, incluyendo simulación de vector de estado, simulación de matriz de densidad y simulación de Monte Carlo.

*   **Ignis**: Ignis es una biblioteca de herramientas para la corrección y mitigación de errores cuánticos. Incluye métodos para medir y caracterizar errores en hardware cuántico, así como técnicas para mitigar esos errores.

*   **Aqua**: Aqua es una biblioteca de algoritmos y aplicaciones cuánticas, como simulaciones de química y problemas de optimización.

*   **IBM Quantum Experience**: IBM Quantum Experience es una plataforma basada en la nube que permite a los usuarios acceder al hardware cuántico de IBM a través de Qiskit. Incluye una interfaz de usuario para diseñar y ejecutar circuitos cuánticos, así como herramientas para analizar los resultados.

El sistema cuántico, a diferencia, se representa por cualquier sistema cuántico de dos niveles; es importante tener en cuenta que el sistema cuántico es probabilístico, es decir, aunque sea un sistema de dos niveles, al medir o intentar saber en cual de los dos niveles está el sistema uno sólo puede obtener una probabilidad ya que el sistema está en todos los estados posibles (superposición) hasta que uno lo mide y lo "Obliga" a escoger un estado; en lenguaje matemático esto lo representamos con la *Esfera de Bloch*.

```python
# Definición del bit clásico
bit = 1
print("Bit Clásico: ", bit)  # Salida o resultado
```

```python
# Representación del qubit usando Qiskit y Aer
from qiskit import QuantumCircuit, execute, Aer

# Crear un circuito cuántico de un qubit (#quantum bits, #quantum bits)
circuit = QuantumCircuit(1, 1)

# Medir el bit cuántico
circuit.measure(0, 0)

# Simulación del circuito en un computador clásico con AER
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, simulator).result()

# Contar los resultados de la medición
counts = result.get_counts(circuit)
print(circuit)
print(counts)  # Output: {'0': 1}
```

#### Ejemplo: Cambiando el estado de un (Qu)bit con un operador o compuerta

En computación clásica, podemos suponer que tenemos el bit $1$ y queremos cambiarlo a su estado contrario, $0$, el operador clásico que realiza este cambio se llama NOT y se simboliza con una barra sobre el bit:

$$Not(0)=\bar{0}=1$$
$$Not(1)=\bar{1}=0$$

Ahora, en computación cuántica existen varios operadores sobre un sólo bit, pero veamos cómo se representa el operador cuántico NOT usando matrices:

$$NOT=X=\sigma_x=\begin{pmatrix}
0 & 1 \\
1 & 0 
\end{pmatrix}$$

es decir, es la matriz de pauli en $x$. Aplicandola sobre un Qubit:

$$X\ket{0}=\begin{pmatrix}
0 & 1 \\
1 & 0 
\end{pmatrix}\begin{pmatrix}
1  \\
0 
\end{pmatrix}$$

$$X\ket{0}=\begin{pmatrix}
0  \\
1 
\end{pmatrix}$$


```python

# Representing a qubit using the Qiskit library
from qiskit import QuantumCircuit, execute, Aer

# Create a quantum circuit with one qubit
circuit = QuantumCircuit(1, 1)

# Apply a Hadamard gate to the qubit
circuit.x(0)

# Measure the qubit
circuit.measure(0, 0)

# Simulate the circuit on a classical computer using the Aer simulator
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, simulator).result()

# Get the counts of the measurements
counts = result.get_counts(circuit)

print(circuit)
print("Bit Cuantico: ", counts)
```

```python
from qiskit import QuantumCircuit, execute, Aer

# Create a quantum circuit with two qubits and two classical bits
circuit = QuantumCircuit(2, 2)

# Apply a NOT gate to the first qubit
circuit.x(0)

# Measure both qubits
circuit.measure([1, 0], [1, 0])

# Simulate the circuit on a classical computer using the Aer simulator
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, simulator).result()

# Get the counts of the measurements
counts = result.get_counts(circuit)

print(circuit)
print(counts)  # Output: {'10': 1024}
```
